<h2> VISUALIZAÇÃO </h2>

<p> Vamos ver sobre pokemons do tipo <b>Fogo X Água</b> </p>

```python


#pokemons do tipo fogo
fogo = pokemon[(pokemon['Type 1'] == 'Fire') | ((pokemon['Type 2']) == 'Fire')]

#pokemons do tipo agua
agua = pokemon[(pokemon['Type 1'] == 'Water')| ((pokemon["Type 2"]) == 'Water')]

#scatter plot
plt.scatter(fogo.Attack.head(50), fogo.Defense.head(50), color='R', label = 'Fogo',marker='*',s=50)

plt.scatter(agua.Attack.head(50), agua.Defense.head(50), color='B', label = 'Agua'          ,s=25)

plt.xlabel("Ataque")
plt.ylabel("Defesa")
plt.legend()
plt.plot()

plt.show()
```

<ul><li>Esse gráfico mostra que pokemons do tipo fogo tem um melhor ataque que os do tipo agua, mas tem menor defesa contra os mesmos.</li></ul>

```python
#Número de pokemons por tipo e por geração

#Type 1

a = pokemon.groupby(['Generation','Type 1']).count().reset_index()
a = a[['Generation','Type 1','Total']]
a = a.pivot('Generation','Type 1','Total')
a[['Water','Fire','Grass','Dragon','Normal','Rock','Flying','Electric']].plot(color= ['b','r','g','#FFA500','brown','#6666ff','#001012','y'],marker='o')
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()

```

<h3>Conclusão Parcial</h3>
<p>Nesse grafico podemos ver que pokemons do tipo agua vem crescendo na primeira geração. Entretanto, o número foi diminuindo
com o passar das gerações. Os do tipo grama tem aumentado desde a segunda até a quinta geração.</p>

```python
a= pokemon.groupby(['Generation','Type 2']).count().reset_index()
a=a[['Generation','Type 2','Total']]
a=a.pivot('Generation','Type 2','Total')
a[['Water','Fire','Grass','Dragon','Normal','Rock','Flying','Electric']].plot(color=['b','r','g','#FFA500','brown','#6666ff','#001012','y'],marker = 'o')
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```

<ul><li>Em relação ao Type 2 o número de pokémons do tipo grama tem aumentando gradativamente, o mesmo pode se dizer para
os tipos dragão</li></ul>

<p> Vamos ver a correlação dos atributos</p>

```python
plt.figure(figsize=(10,4))
sns.heatmap(pokemon.corr(),cbar='viridis',annot=True)
```

<p>Nesse gráfico vemos que os atributos Hp, Attack, Sp. Attack, Sp.Def e Speed estão altamente correlacionados com o atributo Total, sendo o atributo Sp.Atk o maior.</p>

```python
dims = (11.7,8.27) #A4 dimensions
fig, ax = plt.subplots(figsize=dims)
Atthist = sns.distplot(pokemon['Attack'],color='r',hist=False,ax=ax)
SpAhist = sns.distplot(pokemon['Defense'],color = 'b', hist = False,ax=ax)
SpAhist.set(title = 'Distribuição Ataque e Defesa', xlabel = 'Attack:r , Defense:b')
FigHist=SpAhist.get_figure()
FigHist.show()
```

```python
dims = (11.7,8.27) #A4 dimensions
fig, ax = plt.subplots(figsize=dims)
Atthist = sns.distplot(pokemon['Sp. Atk'],color='r',hist=False,ax=ax)
SpAhist = sns.distplot(pokemon['Sp. Def'],color = 'b', hist = False,ax=ax)
SpAhist.set(title = 'Distribuição Sp.Ataque e Sp.Defesa', xlabel = 'Sp.Attack:r , Sp.Defense:b')
FigHist=SpAhist.get_figure()
FigHist.show()
```

<h3>Conclusão Parcial</h3>

<ul><li>Aqui, a distribuição de ambos os atributos (Ataque e Defesa) é semelhante e pode-se sugerir que ambos são inclinados positivamente. Podemos ver que há um final de cauda significativo nas estatísticas de Defesa, em oposição às estatísticas de Ataque, retratando que a estatística de Defesa tem uma variação maior do que a estatística de ataque também.</li>

<li>Já no segundo gráfico, o ataque especial, em azul, tem uma variação maior do que a da defesa especial. Pode-se ver também que Sp.Def mantém o status mais alto dos dois, aproximadamente em 225.</li></ul>
